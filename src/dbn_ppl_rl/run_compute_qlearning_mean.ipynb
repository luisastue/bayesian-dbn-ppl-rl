```python
import json
import numpy as np
import matplotlib.pyplot as plt
from dbn_ppl_rl import *
```

```python
def load_qlearning_results(input_json_path):
    """Load Q-learning results from a JSON file."""
    with open(input_json_path, 'r') as json_file:
        q_learning_results = json.load(json_file)
    return q_learning_results

ql = load_qlearning_results('json/qlearning/q_learning_results.json')
```

```python

def convert_to_means_type(q_learning_results):
    # Extract individual runs (assume each run is a list of episode rewards)
    episode_rewards = q_learning_results["individual_runs"]

    # Calculate mean rewards per episode
    mean_reward = np.mean(episode_rewards, axis=0)

    # Compute smoothed mean with a growing window at the beginning
    smoothed_mean = []
    smoothed_std_low = []
    smoothed_std_high = []

    for i in range(len(mean_reward)):
        window = min(i + 1, 1000)  # Growing window size until 1000
        window_rewards = mean_reward[max(0, i - window + 1):i + 1]
        mean = np.mean(window_rewards)
        std = np.std(window_rewards)
        smoothed_mean.append(mean)
        smoothed_std_low.append(mean - std)
        smoothed_std_high.append(mean + std)

    # Keys corresponding to smoothed values (e.g., episode indices)
    keys_of_smoothed = list(range(len(smoothed_mean)))

    # Downsample smoothed values: [50:length-200][::120]
    start, end = 50, len(smoothed_mean) - 200
    smoothed_mean = smoothed_mean[start:end][::120]
    smoothed_std_low = smoothed_std_low[start:end][::120]
    smoothed_std_high = smoothed_std_high[start:end][::120]
    keys_of_smoothed = keys_of_smoothed[start:end][::120]

    # Prepare the output dictionary
    means_type = {
        "individual_runs": [[float(x) for x in ep] for ep in episode_rewards],
        "mean_rewards": [float(x) for x in mean_reward],
        "smoothed_mean": [float(x) for x in smoothed_mean],
        "smoothed_std_low": [float(x) for x in smoothed_std_low],
        "smoothed_std_high": [float(x) for x in smoothed_std_high],
        "keys_of_smoothed": keys_of_smoothed,
        "name": q_learning_results["name"],
        "info": q_learning_results["info"]
    }

    # Export to JSON file
    with open("json/qlearning/q_learning_results.json", 'w') as json_file:
        json.dump(means_type, json_file, indent=4)

    return means_type


```

```python
convert_to_means_type(ql)
```
